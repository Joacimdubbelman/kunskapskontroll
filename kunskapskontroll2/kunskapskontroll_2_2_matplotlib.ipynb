### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

Explicit way is recommended, with more control and saving in variables.
The implicit way is faster .

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure is a "blueprint" for holding childs.
Axes is attathed to a figure that contains a region for plotting data.
Axis provide ticks and tick labels to provide scales for the data in the Axes.
And artist is what you see on a figure, like Axes and Axis. Most Artists are tied to an Axes (can not be moved/shared to another Axes).

### When plotting in Matplotlib, what is the expected input data type?

numpy.array or masked numpy arrays or objects that can be passed to numpy.asarray

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 100)
y = x**2
fig, ax = plt.subplots()
ax.plot(x, y, 'y^')  #'y' = yellow, ^= arrows
ax.set_title('y = x^2')
ax.set_ylabel('y-axel')
ax.set_xlabel('x-axel')
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, axs = plt.subplots(1, 2, figsize=(9, 3))  #Skapa figur med en rad och tv kolumner med lite extra bredd
axs[0].scatter(random_data_x, random_data_y, color='r')
axs[0].set_title('Scatter Plot', color='r')
axs[1].bar(names, values, color='g')
axs[1].set_title('Bar Plot', color='g')
```

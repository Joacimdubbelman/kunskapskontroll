## Cleaning data

```python
housing = housing_org.dropna()
housing.info()
```

# Visualizations

### Bar plot of the mean number of households for each ocean proximity

```python
# Calculate the mean number of households for each ocean_proximity category
households_by_proximity = housing.groupby('ocean_proximity')['households'].mean().reset_index()

# Generate random colors for each bar
num_categories = len(households_by_proximity)
random_colors = np.random.rand(num_categories, 3) 

# Plot and labels
plt.figure(figsize=(10, 6))
bars = plt.bar(households_by_proximity['ocean_proximity'], households_by_proximity['households'], color=random_colors)
plt.xlabel('Ocean Proximity')
plt.ylabel('Average Households')
plt.title('Average Number of Households by Ocean Proximity', fontsize=15)
plt.xticks(rotation=45)

# Add values on top of bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2.,  
        height + 5,                         
        f'{int(height)}',                   
        ha='center',                       
        va='bottom'                        
    )
```

### Comments on "Average Number of Households by Ocean Proximity"
* The number of households is fairly evenly distributed in proximity to the sea. The exception is the island, which has fewer households.

## Scatter plot of "Comparison of 1000 Least/Most Expensive Housing Areas"

```python
import matplotlib.image as mpimg

# Sort the housing data by median_house_value
sorted_housing = housing.sort_values(by='median_house_value')

# Select areas
least_expensive_1000 = sorted_housing.head(1000)
most_expensive_1000 = sorted_housing.tail(1000)
three_cheapest = housing.nsmallest(3, 'median_house_value')

plt.figure(figsize=(14, 10))

# Plot least expensive areas in red
plt.scatter(
    x=least_expensive_1000["longitude"],
    y=least_expensive_1000["latitude"],
    alpha=0.6,
    s=30,
    color='red',
    label=f'Least Expensive (${int(least_expensive_1000["median_house_value"].mean()):,})',
    edgecolor='darkred',
    linewidth=0.5
)

# Plot most expensive areas in green
plt.scatter(
    x=most_expensive_1000["longitude"],
    y=most_expensive_1000["latitude"],
    alpha=0.6,
    s=30,
    color='green',
    label=f'Most Expensive (${int(most_expensive_1000["median_house_value"].mean()):,})',
    edgecolor='darkgreen',
    linewidth=0.5
)

# Show California map (Antonios map)
california_img=mpimg.imread('california.png')
plt.imshow(california_img, extent=[-124.55, -113.80, 32.45, 42.05], alpha=0.3)

# Add arrows pointing to the "Lowest median house value"
for i, row in three_cheapest.iterrows():
    plt.annotate('Lowest median house value', 
                 xy=(row['longitude'], row['latitude']),
                 xytext=(row['longitude']-1, row['latitude']+1),
                 arrowprops=dict(facecolor='blue', shrink=0.03, width=2),
                 fontsize=11,
                 color='blue')

# Labels, title and grids
plt.xlabel('Longitude', fontsize=13)
plt.ylabel('Latitude', fontsize=13)
plt.title('Comparison of 1000 Least/Most Expensive Housing Areas', fontsize=15)
plt.legend(loc="best", fontsize=13)
plt.grid(alpha=0.9)
```

### Comments on "Comparison of 1000 Least/Most Expensive Housing Areas"
* A majority of the highest valued houses are close too the sea.

## Circle diagram of "Distribution of Total Rooms by Area"

```python
# Group by ocean_proximity and calculate total_rooms and % in areas
rooms_by_location = housing.groupby('ocean_proximity')['total_rooms'].sum()
total = rooms_by_location.sum()
rooms_percentage = (rooms_by_location / total * 100).round(1)

# Create pie chart
plt.figure(figsize=(10, 8))
plt.pie(rooms_by_location, 
        labels=[f"{loc}\n({pct}%)" for loc, pct in zip(rooms_by_location.index, rooms_percentage)],
        autopct='%1.1f%%',
        startangle=90,
        shadow=True,
        explode=[0.05] * len(rooms_by_location),  # Slight separation for slices
        colors=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FF99CC']) 

plt.title('Distribution of Total Rooms by Area', fontsize=16)
plt.tight_layout()
```

### Comments on "Distribution of Total Rooms by Area"
* If we are a construction company working with private houses, we could investigate which areas to focus on. For example, we would ignore the island and work within 1 hour of the coast. Of course, after competitive analysis and logistics analysis.
